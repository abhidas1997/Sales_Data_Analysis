# Exploraotry Data Analysis

Gender

```python
data.columns
```

```python
#Count of Different gender
gen = sns.countplot(x = 'Gender',hue='Gender', data = data)

for bars in gen.containers:  #Create values of each bar
    gen.bar_label(bars)
```

```python
#Total purchase amount based on paticular geneder

gen_amount = data.groupby(["Gender"], as_index = False)['Amount'].sum()
plt_amount = sns.barplot(x="Gender",y='Amount',data=gen_amount)
for bars in plt_amount.containers:
    plt_amount.bar_label(bars)
```

Age

```python
#Plot on basis of Age
age = sns.countplot(x="Age Group", data = data)

for bars in age.containers:
    age.bar_label(bars)
```

```python
#Plot on basis of Age and Gender
age_gen = sns.countplot(x="Age Group", data = data, hue = "Gender")

for bars in age_gen.containers:
    age_gen.bar_label(bars)
```

```python
#Total Amount vs Age
age_amount = data.groupby(["Age Group"], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt_amount = sns.barplot(x="Age Group",y='Amount',data=age_amount)
```

State

```python
#State Vs Orders
state_amount = data.groupby(["State"], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(23,5)})
plt_amount = sns.barplot(x="State",y='Orders',data=state_amount)
for bars in plt_amount.containers:
    plt_amount.bar_label(bars)
```

```python
#State Vs Amount
state_amount = data.groupby(["State"], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(23,5)})
plt_amount = sns.barplot(x="State",y='Amount',data=state_amount)
for bars in plt_amount.containers:
    plt_amount.bar_label(bars)
```

Marital_Status

```python
#Count of Marital_Status
marital_status = sns.countplot(x = 'Marital_Status', data = data)
sns.set(rc={'figure.figsize':(10,8)})
for bars in marital_status.containers:  #Create values of each bar
    marital_status.bar_label(bars)
```

```python
#Count of Marital_Status based on Gender
marital_status_gen = sns.countplot(x = 'Marital_Status',hue='Gender', data = data)
sns.set(rc={'figure.figsize':(20,5)})
for bars in marital_status_gen.containers:  #Create values of each bar
    marital_status_gen.bar_label(bars)
```

```python
#Count of Purchasing based on Maitial Status
marital_status_amount = data.groupby(["Marital_Status"], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
plt_maritial_status = sns.barplot(x="Marital_Status", y='Amount', data = marital_status_amount)
for bars in plt_maritial_status.containers:  #Create values of each bar
    plt_maritial_status.bar_label(bars)
```

Occupation

```python
#Plot on basis of Occupation
jobs = sns.countplot(x="Occupation", data = data)
sns.set(rc={'figure.figsize':(40,5)})
for bars in jobs.containers:
    jobs.bar_label(bars)
```

```python
#Count of Purchasing based on Occupation
jobs_amount = data.groupby(["Occupation"], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
jobs_status = sns.barplot(x="Occupation", y='Amount', data = jobs_amount)
# for bars in jobs_status.containers:  #Create values of each bar
#     jobs_status.bar_label(bars)
```

Product Category

```python
#Plot on basis of Product
prodct = sns.countplot(x="Product_Category", data = data)
sns.set(rc={'figure.figsize':(40,10)})
for bars in prodct.containers:
    prodct.bar_label(bars)
# cal = data.groupby(["Product_Category"])['Amount'].sum()
# plt.pie(cal)
```

```python
#Count of Amount based on Product Category
pro_amount = data.groupby(["Product_Category"], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
pro_status = sns.barplot(x="Product_Category", y='Amount', data = pro_amount)
```

```python
#Count of Amount based on Product Category
pro_id_amount = data.groupby(["Product_ID"], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
pro_id_status = sns.barplot(x="Product_ID", y='Amount', data = pro_id_amount)
```
